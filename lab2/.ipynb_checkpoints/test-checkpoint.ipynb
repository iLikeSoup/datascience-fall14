```python
join = pd.merge(players, countries, on='country_id')
join_df = pd.DataFrame(join)
new_col = join_df.columns.values
new_col[4] = 'country_name'
join_df.columns = new_col

groupby_series = join_df.groupby(['country_name']).size()
groupby_df = pd.DataFrame(groupby_series)
groupby_df.columns = ['num_players']
groupby_df
```

```python
join = pd.merge(players, countries, on='country_id')
df_join = pd.DataFrame(join)
t = []

for i in range(len(df_join.index)):
    year = df_join['birthdate'][i][:4]
    if year == '1975':
        t.append(df_join['country_id'][i])
    
t_df = pd.DataFrame(t, columns = ['country_id'])
t2_df = pd.DataFrame(t_df.groupby('country_id').size())
c_df = pd.DataFrame(countries)

key = []
val = []
index = []
count = 0

for i in t2_df.index:
    key.append(i)
    val.append(t2_df.loc[i][0])
    index.append(count)
    count += 1

d = {'country_id' : pd.Series(key, index=index),
     'num_players' : pd.Series(val, index=index)}
t3_df = pd.DataFrame(d)

b = pd.merge(c_df, t3_df, on ='country_id', how='left')
for i in range(len(b.index)):
    if str(b['num_players'][i]) == 'nan':
        b['num_players'][i] = 0
b[['country_id', 'num_players']]
```
