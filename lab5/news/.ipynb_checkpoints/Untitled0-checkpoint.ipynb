```python
import nltk
import re
```

```python
patt = re.compile('([A-Z]+) ([\w]+)\/NNP')

with open("mn1.html", "r") as myfile:
    data = myfile.read()
sentences = nltk.sent_tokenize(data)
sentences = [nltk.word_tokenize(sent) for sent in sentences]
sentences = [nltk.pos_tag(sent) for sent in sentences]
chunk = [nltk.ne_chunk(sent) for sent in sentences]
for c in chunk:
    regex = patt.findall(str(c))
    if regex:
        for r in regex:
            print('{}, {}'.format(r[1], r[0]))
```

```python
IN = re.compile(r'.*\bin\b(?!\b.+ing)')
for doc in nltk.corpus.ieer.parsed_docs('NYT_19980315'):
    for rel in nltk.sem.extract_rels('ORG', 'LOC', doc, corpus='ieer', pattern = IN):
        print(nltk.sem.rtuple(rel))
```

```python
IN = re.compile(r'.*\b(in|for|of|at)\b')
for doc in nltk.corpus.ieer.parsed_docs('NYT_19980315'):
    for rel in nltk.sem.extract_rels('PER', 'ORG', doc, corpus='ieer', pattern = IN):
        print(nltk.sem.rtuple(rel))
```
